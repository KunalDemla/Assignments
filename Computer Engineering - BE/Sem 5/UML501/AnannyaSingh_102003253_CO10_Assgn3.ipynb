```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML_Lab/USA_Housing.csv', sep=',')
df
```

```python
X = df.iloc[:,0:5]
X
```

```python
Y = df.iloc[:,5]
```

```python
Y = np.array(Y)
print(Y.shape)
Y = Y.reshape(-1,1)
Y.shape
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = np.insert(X_scaled,0,1,axis=1)
X_scaled.shape
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
kfold = KFold(n_splits = 5, random_state = 1, shuffle = True)

betas = []
score = []
predicts = []

pred_vals = []
i=1
for train_indice, test_indice in kfold.split(X_scaled):
  
  X_train , X_test = X_scaled[train_indice], X_scaled[test_indice]
  Y_train , Y_test = Y[train_indice], Y[test_indice]

  A = X_train.T.dot(X_train)  #X_transpose*X
  B = np.linalg.inv(A)
  C = B.dot(X_train.T) #(X_tranpose*X)invers*X
  beta = C.dot(Y_train)
  
  betas.append(beta)

  y_predict = X_test.dot(beta)
  predicts.append(y_predict)
  err = r2_score(y_predict,Y_test)
  print(f'For iteraton {i}: {err} ')

  score.append(err)
  i = i+1



  
```

```python
maxVal = max(score)
maxIdx = score.index(maxVal)
maxIdx

bestBeta = betas[maxIdx]
bestBeta
```

Q2 Gradient Descent

```python
df = pd.read_csv('/content/drive/MyDrive/ML_Lab/USA_Housing.csv', sep=',')
df
```

```python
X = df.iloc[:,0:5]
X
```

```python
Y = df.iloc[:,5]
```

```python
Y = np.array(Y)
print(Y.shape)
Y = Y.reshape(-1,1)
Y.shape
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = np.insert(X_scaled,0,1,axis=1)
X_scaled.shape
```

```python
from sklearn.model_selection import train_test_split

train_ratio = 0.56
validation_ratio = 0.14
test_ratio = 0.30

# train is now 56% of the entire data set
# the _junk suffix means that we drop that variable completely
x_train, x_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=1-train_ratio)

# test is now 30% of the initial data set
# validation is now 14% of the initial data set
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=test_ratio/(test_ratio + validation_ratio)) 
```

```python
learning_rate = [0.001, 0.01, 0.1, 1]
```

```python
from sklearn.base import BaseEstimator
beta = np.zeros(6)
num_iter = 1000
n = len(x_train)
max_beta = None
max_r2_score = -100000


for i in learning_rate:
    for j in range(1000):

        x0_grad = 0
        x1_grad = 0
        x2_grad = 0
        x3_grad = 0
        x4_grad = 0
        x5_grad = 0

        for k in range(len(x_train)):

            a = x_train[k,0]
            b = x_train[k,1]
            c = x_train[k,2]
            d = x_train[k,3]
            e = x_train[k,4]
            f = y_train[k,]
            # print(f)
            
            x0_grad += (beta[0] + (beta[1]*a) + (beta[2]*b) + (beta[3]*c) + (beta[4]*d) + (beta[5]*e) - f)
            x1_grad += ((beta[0] + (beta[1]*a) + (beta[2]*b) + (beta[3]*c) + (beta[4]*d) + (beta[5]*e) - f) * a)
            x2_grad += ((beta[0] + (beta[1]*a) + (beta[2]*b) + (beta[3]*c) + (beta[4]*d) + (beta[5]*e) - f) * b)
            x3_grad += ((beta[0] + (beta[1]*a) + (beta[2]*b) + (beta[3]*c) + (beta[4]*d) + (beta[5]*e) - f) * c) 
            x4_grad += ((beta[0] + (beta[1]*a) + (beta[2]*b) + (beta[3]*c) + (beta[4]*d) + (beta[5]*e) - f) * d)
            x5_grad += ((beta[0] + (beta[1]*a) + (beta[2]*b) + (beta[3]*c) + (beta[4]*d) + (beta[5]*e) - f) * e)

        beta[0] = beta[0] - i / n * x0_grad
        beta[1] = beta[1] - i / n * x1_grad
        beta[2] = beta[2] - i / n * x2_grad
        beta[3] = beta[3] - i / n * x3_grad
        beta[4] = beta[4] - i / n * x4_grad
        beta[5] = beta[5] - i / n * x5_grad
        
        beta = np.array(beta).reshape(-1,1)
        y_pred = x_val.dot(beta)

        error = y_val - y_pred
        sq_err = np.power(error, 2)

        sum_sq_err = np.sum(sq_err)
        mse = sum_sq_err/len(y_pred)
        rmse = np.sqrt(mse)

        y_mean = np.mean(y_val)
        total_variance = np.sum((y_val-y_mean)**2)

        r2_score = (1-sum_sq_err/total_variance)
        # print(r2_score)

        if r2_score > max_r2_score:
            best_learning_rate = i
            max_beta = beta
            max_r2_score = r2_score




```

```python
print(max_beta)
print(best_learning_rate)
```

```python

y_pred = x_test.dot(max_beta)
error = y_test - y_pred
sq_err = np.power(error, 2)

sum_sq_err = np.sum(sq_err)
mse = sum_sq_err/len(y_pred)
rmse = np.sqrt(mse)

y_mean = np.mean(y_test)
total_variance = np.sum((y_test-y_mean)**2)

r2_score = (1-sum_sq_err/total_variance)
```

```python
print(r2_score)
```
